```python
x = iris.drop('species', axis =1)
```

```python
y = iris['species']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.7,random_state=1)
```

```python
from sklearn.metrics import confusion_matrix, recall_score, precision_score, f1_score, roc_auc_score,accuracy_score
```

## Random Forest Classifier



```python
from sklearn.ensemble import RandomForestClassifier
rfcl = RandomForestClassifier(n_estimators = 50, random_state=42,max_features=7)
rfcl = rfcl.fit(x_train, y_train)
```

```python
y_predict = rfcl.predict(x_test)
```

```python
y_predict
```

```python
print("Training accuracy",rfcl.score(x_train,y_train))
print()
print("Testing accuracy", rfcl.score(x_test, y_test))
```

## XGB Boost Classifier

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb = xgb.fit(x_train, y_train)
y_predict = xgb.predict(x_test)
```

```python
print("Training accuracy",xgb.score(x_train,y_train))
print()
print("Testing accuracy", xgb.score(x_test, y_test))
```

## Both models work really well, probably as this is a very simple dataset.
